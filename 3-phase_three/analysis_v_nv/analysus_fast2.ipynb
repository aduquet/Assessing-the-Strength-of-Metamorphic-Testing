```python
final_results = []
for paths in target_paths:
    
    mutated_file_name = paths.split('/')[-1]
    # print(mutated_file_name)
    target_MR1_file = os.path.join(paths, 'final_MR1_' + mutated_file_name + '.json')
    problems = []
    try:
        with open(target_MR1_file, 'r') as file:
            mr1 = json.load(file)
        cnt_v = 0
        cnt_no_v = 0
        otro = 0
        status = []
        for key, value in mr1.items():
            try:
                if value['MR1_status'] == "no-violated":
                    cnt_no_v += 1
                
                elif value['MR1_status'] == "violated":
                    cnt_v += 1
                else:
                    otro += 1
                # status.append('All good')
            except :
                status.append('ID NOT FOUNT/PROBLEMS ->' + key)
        
        final_file = {
            'file_name': mutated_file_name, 
            'cnt_no_v': cnt_no_v,
            'cnt_v': cnt_v,
            'otro': otro,
            'status': status
            }
        
        final_results.append(final_file)
        
    except :
        problems.append('File not found ->  ' + mutated_file_name)
df_final_results = pd.DataFrame(final_results)
# Save the DataFrame to a CSV file
df_final_results.to_csv('analysis_v_nv_MR1.csv')
# with open('analysis_v_nv_MR1.json', 'w') as json_file:
#     json.dump(final_results, json_file, indent=4)
```
