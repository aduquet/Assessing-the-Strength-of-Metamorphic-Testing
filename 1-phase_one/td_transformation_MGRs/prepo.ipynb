```python
'''MR_MUL'''

with open(test_data_path, 'r') as file:
    test_data = json.load(file)
    
for key, value in test_data.items():
    const = int(np.random.randint(low=2, high=6, size=1))
    test_data[key]['const'] = const
    test_data[key]['ttd_xdat'] = list(np.array(value['td_xdat']) * const)
    test_data[key]['ttd_ydat'] = list(np.array(value['td_ydat']) * const)
```

```python
def MR_INV(td_a, td_b):
    # MR_INV -> Take the inverse of each element
    aux_list_a = td_a.copy()
    aux_list_b = td_b.copy()
    
    if len(td_a) != 0 and type(td_a) == list and len(td_b) != 0 and type(td_b) == list:
        for item in td_a:
            if item != 0:
                aux_list_a.append(1 / item )

            else:
                aux_list_a.append('ZeroDivisionError')
                
        for item in td_b:
            if item != 0:
                aux_list_b.append(1 / item )

            else:
                aux_list_b.append('ZeroDivisionError')
        
    return aux_list_a, aux_list_b
```

```python
def MRduplicating(x,y):
    aux_list_a = x.copy()
    aux_list_b = y.copy()
    if(len(x)==16):
        
        return aux_list_a, aux_list_b
    
    else:
        if(len(x) < 10):

            duplicated = int(np.random.randint(low=1, high=5, size=1))
            
            for i in range(0,duplicated):
                position = np.random.randint(0, len(x) -1)
                dupl_x = x[position]
                dupl_y = y[position]
                
                aux_list_a.append(dupl_x)
                aux_list_b.append(dupl_y)
            
            return aux_list_a, aux_list_b

        else:
            
            position = np.random.randint(0, len(x) -1)
            dupl_x = x[position]
            dupl_y = y[position]
            aux_list_a.append(dupl_x)
            aux_list_b.append(dupl_y)
            
            return aux_list_a, aux_list_b
```

```python

with open(test_data_path, 'r') as file:
    test_data = json.load(file)
    
for key, value in test_data.items():
    const = int(np.random.randint(low=2, high=6, size=1))
    position = np.random.randint(0, len(value['td_xdat'])-1)
    aux_list_x = value['td_xdat'].copy()
    aux_list_y = value['td_ydat'].copy()
    aux_list_x2 = value['td_xdat'].copy()
    aux_list_y2 = value['td_ydat'].copy()
    aux_list_x.pop(position)
    aux_list_y.pop(position)
    position = np.random.randint(0, len(value['td_xdat']) -1)
    aux_list_x2.insert(int(position), int(const))
    aux_list_y2.insert(int(position), int(const))
    
    invx, invy = MR_INV(value['td_xdat'], value['td_ydat'])
    
    dupli_x, dupli_y = MRduplicating(value['td_xdat'], value['td_ydat'])
    
    test_data[key]['const'] = int(const)
    test_data[key]['ttd_xdat_MRG1_mul'] = [x * const for x in value['td_xdat']] 
    test_data[key]['ttd_ydat_MRG1_mul'] = [y * const for y in value['td_ydat']] 
    test_data[key]['ttd_xdat_MRG2_exc'] = aux_list_x
    test_data[key]['ttd_ydat_MRG2_exc'] = aux_list_y
    test_data[key]['ttd_xdat_MRG3_inc'] = aux_list_x2
    test_data[key]['ttd_ydat_MRG3_inc'] = aux_list_y2
    test_data[key]['ttd_xdat_MRG4_inv'] = invx
    test_data[key]['ttd_ydat_MRG4_inv'] = invy
    test_data[key]['ttd_xdat_MRG5_reflection'] = [-xi for xi in value['td_xdat']]
    test_data[key]['ttd_ydat_MRG5_reflection'] = [-yi for yi in value['td_ydat']]
    test_data[key]['ttd_xdat_MRG6_duplicate'] = dupli_x
    test_data[key]['ttd_ydat_MRG6_duplicate'] = dupli_y

with open('test_data_MRG.json', 'w') as outfile:
    json.dump(test_data, outfile, indent=4) 
```
